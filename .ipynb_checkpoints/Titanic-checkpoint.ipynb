```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import metrics 
import warnings
warnings.simplefilter("ignore")
df = pd.DataFrame(pd.read_csv('train.csv'))
df["Name_Length"] = df["Name"].apply(len)
display(df)
```

#### preprocess_df adds the missing values and makes categorical values oridnal values to the raw dataframes

```python

def preprocess_df(dataframe):
    dataframe.replace({'Sex': {'male': 0, 'female' : 1}}, inplace=True)
    dataframe.replace({'Embarked': {'S': 0, 'C' : 1, 'Q' : 2, float('nan'): -1}}, inplace=True)
    dataframe.replace({'Cabin': {float('nan') : 'O00'}}, inplace=True)
    dataframe.replace({'Age': {float('nan') : -1}}, inplace=True)
    dataframe.replace({'Fare': {float('nan') : -1}}, inplace=True)
    dataframe.Sex = dataframe.Sex.astype(int)
    dataframe.Embarked = dataframe.Embarked.astype(int)
```

#### prob_plots compares the probability curves of the 2 distributions

```python
def prob_plots(feature_str):
    
    column = pd.DataFrame(df[feature_str])
    list1 = column[df["Survived"] == 1][feature_str].value_counts()/len(column[df["Survived"] == 1])
    list2 = column[df["Survived"] == 0][feature_str].value_counts()/len(column[df["Survived"] == 0])


    fig = plt.figure(figsize = (16,10))


    plt.title("Comparison of probability distributions of the " + feature_str)
    plt.bar(
            x = range(len(list1)),
            height = [list1[i] for i in range(len(list1))],
            color = "green",
            alpha = 0.3
    )
    plt.bar(
            x = range(len(list2)),
            height = [list2[i] for i in range(len(list2))],
            color = "red",
            alpha = 0.3
    )
```

#### Baseline Classifier

```python
print(X_train.describe())
```

```python
NUM_TESTS = 3000

def test_estimator(test_features, num_tests):
    f1_no = list()
    f1_yes = list()
    display(test_features)
    for i in range(num_tests):
        Estimator = RandomForestClassifier()
        X_prelim_train, X_prelim_test, y_prelim_train, y_prelim_test = train_test_split(test_features, y_train, test_size = 0.15)
        Estimator.fit(X_prelim_train, y_prelim_train) 
        Estimator_predictions = Estimator.predict(X_prelim_test)
        prec, rec, f1, sup = metrics.precision_recall_fscore_support(y_prelim_test, Estimator_predictions)
        no, yes = f1[0], f1[1]
        f1_no.append(no); f1_yes.append(yes)

    return f1_no, f1_yes
            
def paired_hist(plt1, plt2, c1, c2):
    sns.distplot(plt1, bins = 10, color = c1)
    sns.distplot(plt2, bins = 10, color = c2)
```

```python
train = X_train.drop(columns = ["Name_Length"])
no, yes = test_estimator(train, NUM_TESTS)
no2, yes2 = test_estimator(X_train, NUM_TESTS)
```

```python
paired_hist(no, yes, "green", "red")
paired_hist(no2, yes2, "yellow", "purple")
print("No Comparison")
print(pd.DataFrame(no).describe())
print(pd.DataFrame(no2).describe())
print("\nYes Comparison")
print(pd.DataFrame(yes).describe())
print(pd.DataFrame(yes2).describe())
```

## We can see from the figures above that the inclusion of the name_length composite attribute does not affect the classifier, we turn to more preprocessing

We can start by adding the cabin levels, where the first letter of the cabin is the level, and compare the probability distributions of the cabin levels in the surviving group vs the non-surviving group

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

first_char = lambda x : x[0]
cabin_levels = df['Cabin'].apply(first_char)
cabin_levels_ordinal = pd.DataFrame(le.fit_transform(cabin_levels))

display(cabin_levels)
display(cabin_levels_ordinal)

prob_plots(cabin_levels_ordinal, 'cabin levels')

```
